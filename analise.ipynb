```python
import pandas as pd
import IPython.display as display

nome_arquivo = "vendas_biscoitos.csv"
df = pd.read_csv(nome_arquivo)

df['data'] = pd.to_datetime(df['data'])
```

```python
# Ranking de locais campeões de vendas de cada mês

df['mes_ano'] = df['data'].dt.to_period('M')

ranking_locais = df.groupby(['local_venda', 'mes_ano']).agg(
    total_pedidos=('numero_pedido', 'nunique'),
    preco_total=('preco', 'sum')
).reset_index()

ranking_locais = ranking_locais.sort_values(by=['mes_ano', 'preco_total'], ascending=[True, False])

import IPython.display as display
display.display(ranking_locais)
```

```python
# Ranking de biscoitos campeões de vendas de cada local

ranking_biscoitos = df.groupby(['local_venda', 'nome_biscoito']).agg(
    total_vendido=('preco', 'sum'),
    total_pedidos=('numero_pedido', 'nunique')
).reset_index()

ranking_biscoitos = ranking_biscoitos.sort_values(by=['local_venda', 'total_vendido'], ascending=[True, False])

ranking_top_biscoitos = ranking_biscoitos.groupby('local_venda').head(1)

display.display(ranking_top_biscoitos)
```

```python
# Ranking de biscoitos campeões de vendas de cada local por mês

df['mes_ano'] = df['data'].dt.to_period('M')

ranking_biscoitos = df.groupby(['local_venda', 'mes_ano', 'nome_biscoito']).agg(
    total_vendido=('preco', 'sum'),
    total_pedidos=('numero_pedido', 'nunique')
).reset_index()

ranking_biscoitos = ranking_biscoitos.sort_values(by=['local_venda', 'mes_ano', 'total_vendido'], ascending=[True, True, False])

ranking_top_biscoitos = ranking_biscoitos.groupby(['local_venda', 'mes_ano']).head(1)

display.display(ranking_top_biscoitos)
```

```python
# Ranking de biscoitos campeões de vendas de todos os locais

ranking_biscoito_geral = df.groupby('nome_biscoito').agg(
    total_vendido=('preco', 'sum'),
    total_pedidos=('numero_pedido', 'nunique')
).reset_index()

ranking_biscoito_geral = ranking_biscoito_geral.sort_values(by='total_vendido', ascending=False)

biscoito_mais_vendido = ranking_biscoito_geral.head(5)

display.display(biscoito_mais_vendido)
```
